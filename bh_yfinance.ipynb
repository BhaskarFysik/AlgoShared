```python
import yfinance as yf
symbol='ITC'
df = yf.Ticker(f"{symbol}.NS")
```

```python
# get stock info
stock_info=df.info
stock_info
```

```python
# get historical market data
stock_hist = df.history(period="60d",interval="15m")
stock_hist
```

```python
# show actions (dividends, splits)
stock_actions=df.actions
stock_actions
```

```python
# show dividends
stock_dividends=df.dividends
stock_dividends
```

```python
# show splits
stock_splits=df.splits
stock_splits
```

```python
# show financials
stock_financials=df.financials
stock_quarterly_financials=df.quarterly_financials
stock_financials
```

```python
# show major holders
stock_majorholding=df.major_holders
stock_majorholding
```

```python
# show institutional holders
stock_inst_holding=df.institutional_holders
stock_inst_holding
```

```python
# show balance sheet
stock_balance_sheet=df.balance_sheet
stock_quarterly_balance_sheet=df.quarterly_balance_sheet
stock_balance_sheet
```

```python
# show cashflow
stock_cashflow=df.cashflow
stock_quarterly_cashflow=df.quarterly_cashflow
stock_cashflow
```

```python
# show earnings
stock_earnings=df.earnings
stock_quarterly_earnings=df.quarterly_earnings
stock_earnings
```

```python
# show sustainability
stock_sustainability=df.sustainability
stock_sustainability
```

```python
# show analysts recommendations
stock_recommendations=df.recommendations
stock_recommendations
```

```python
# show next event (earnings, etc)
stock_events=df.calendar
stock_events
```

```python
# show all earnings dates
stock_earnings_dates=df.earnings_dates
stock_earnings_dates
```

```python
# show ISIN code - *experimental*
# ISIN = International Securities Identification Number
stock_isin=df.isin
stock_isin
```

```python
# show news
stock_news=df.news
stock_news
```
